•	Load the Titanic dataset (train.csv) (download link: https://www.kaggle.com/competitions/titanic/data) into a Pandas DataFrame.

```python
import pandas as pd
data = pd.read_csv("train.csv")
data
```

o	Basic statistics (.describe())

```python
data.describe()
```

o	Data types (.dtypes)

```python
data.dtypes
```

o	Number of unique values per column (.nunique())

```python
data.nunique()
```

•	Extract all passengers who traveled in 2nd class.

```python
second_class_passengers = data[data["Pclass"] == 2]
second_class_passengers
```

•	Retrieve the name and fare of all passengers who paid more than $100.

```python
high_fare_passengers = data[data["Fare"] > 100][["Name", "Fare"]]
high_fare_passengers
```

•	Extract passengers who are older than 30 and traveled in 1st class.

```python
first_class_older_passengers = data[(data["Age"] > 30) & (data["Pclass"] == 1)]
first_class_older_passengers 
```

•	Select passengers who embarked from 'C' and survived.

```python
embarked_c_survived = data[(data["Embarked"] == "C") & (data["Survived"] == 1)]
embarked_c_survived
```

•	Find the average fare paid per class (Pclass).

```python
average_fare_per_class = data.groupby("Pclass")["Fare"].mean()
average_fare_per_class
```

•	Count the number of survivors by gender.

```python
survivors_by_gender = data[data["Survived"] == 1]["Sex"].value_counts()
survivors_by_gender
```

•	Find the median age of passengers grouped by embarkation port.

```python
median_age_by_embarkation = data.groupby("Embarked")["Age"].median()
median_age_by_embarkation
```

•	Create a new column "FamilySize" as the sum of SibSp and Parch plus one (the passenger themselves).

```python
data["FamilySize"] = data["SibSp"] + data["Parch"] + 1
data
```

•	Create a new column "FarePerPerson" by dividing "Fare" by "FamilySize".

```python
data["FarePerPerson"] = data["Fare"] / data["FamilySize"]
data
```

•	Create a new column "AgeGroup": \
o	"Child" if Age < 18 \
o	"Adult" if 18 ≤ Age < 60 \
o	"Senior" if Age ≥ 60


```python
data["AgeGroup"] = data["Age"].apply(lambda x: "Unknown" if pd.isna(x) else "Child" if x < 18 else "Adult" if x < 60 else "Senior")
data
```

•	Create two DataFrames (df1 and df2) from the Titanic dataset: \
o	df1 → Includes PassengerId, Name, Pclass, and Survived \
o	df2 → Includes PassengerId, Age, Fare, FamilySize, and Embarked


```python
df1 = data[["PassengerId", "Name", "Pclass", "Survived"]]

df2 = data[["PassengerId", "Age", "Fare", "FamilySize", "Embarked"]]
```

```python
df1
```

```python
df2
```

•	Merge the two DataFrames on PassengerId.

```python
merged_df = pd.merge(df1, df2, on="PassengerId")
merged_df
```

•	Missing values count (.isna().sum())

```python
data.isna().sum()
```

•	Fill missing "Age" values with the median age.

```python
data["Age"].fillna(data["Age"].median(), inplace=True)
data
```

```python
data.isna().sum()
```

•	Fill missing "Embarked" values with the most frequent embarkation port.

```python
data["Embarked"].fillna(data["Embarked"].mode()[0], inplace=True)
```

```python
data.isna().sum()
```
