```python
import pandas as pd
data = pd.read_csv("train.csv")
data
```

Drop unnecessary columns (PassengerId, Name, Parch, Ticket, Cabin)

```python
data = data.drop(columns=["PassengerId", "Name", "Parch", "Ticket", "Cabin"])
data
```

Missing values count

```python
data.isna().sum()
```

•	Fill missing values in the age column with the median age

```python
data["Age"].fillna(data["Age"].median(), inplace=True)
data
```

```python
data.isna().sum()
```

•	Fill missing values in the embarked column with the most frequent value

```python
data["Embarked"].fillna(data["Embarked"].mode()[0], inplace=True)
```

```python
data.isna().sum()
```

```python
data
```

•	Convert sex into binary (male = 0, female = 1)

```python
data["Sex"] = data["Sex"].map({"male": 0, "female": 1})
data
```

•	Convert embarked into numerical values (C=0, Q=1, S=2)

```python
data["Embarked"] = data["Embarked"].map({"C": 0, "Q": 1, "S": 2})
data
```

•	Convert age into age groups (Child, Teenager, Adult, Senior) 

```python
data["AgeGroup"] = data["Age"].apply(lambda x: "Unknown" if pd.isna(x) else "Child" if x < 12 else "Teenager" if 12 < x < 18 else "Adult" if x < 60 else "Senior")
with pd.option_context('display.max_rows', None, 'display.max_columns', data.shape[1]):
    print(data)
```

•	Histogram: Plot the distribution of age groups using sns.count plot

```python
data["AgeGroup"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.countplot(x=data["AgeGroup"], order=["Child", "Teenager", "Adult", "Senior"])
plt.title("Distribution of Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

•	Bar plot: Use a bar plot to show survival rates (Survived) across age groups.

```python
data[data["Survived"] == 1]["AgeGroup"].value_counts()
```

```python
survival_rates_age = data.groupby("AgeGroup")["Survived"].mean()

plt.figure(figsize=(8, 6))
ax = sns.barplot(x=survival_rates_age.index, y=survival_rates_age.values, order=["Child", "Teenager", "Adult", "Senior"])

for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title("Survival Rates Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.show()
```

•	Count Plot: Visualize the number of passengers per class (Pclass), differentiating by (hue) age group. Use “husl” palette 

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(x=data["Pclass"], hue=data["AgeGroup"], palette="husl")


for p in ax.patches:
    height = p.get_height()
    if height > 0:  
        ax.annotate(f'{int(height)}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.title("Number of Passengers per Class by Age Group")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title="Age Group")
plt.show()
```

•	Boxplot: compare fare distribution across different classes (Pclass). 

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=data["Pclass"], y=data["Fare"], palette="husl")
plt.title("Fare Distribution Across Different Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

•	Scatter Plot: plot Age vs. Fare, differentiating by (hue) survival status (Survived). Use a palette of your own choice.

```python
custom_palette = {0: "red", 1: "green"}

plt.figure(figsize=(10, 6))
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=data, palette=custom_palette, s=100)

plt.title("Scatter Plot of Age vs. Fare (Colored by Survival Status)", fontsize=16)
plt.xlabel("Age", fontsize=14)
plt.ylabel("Fare", fontsize=14)


plt.legend(title="Survived?", loc="upper right")
plt.show()
```

•	Violin Plot: analyze age distribution within survival categories.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x="Survived", y="Age", data=data, palette="Set2", split=False)

plt.title("Age Distribution by Survival Status", fontsize=16)
plt.xlabel("Survived", fontsize=14)
plt.ylabel("Age", fontsize=14)

plt.show()
```

•	Heatmap (with pre-processing step): select only Age, Fare, Sex, Pclass, SibSp, and Survived columns and generate a correlational heatmap. [use sns.heatmap() function and inside call the dataframe with .corr() function for correlational heatmap] 

```python
data_selected = data[["Age", "Fare", "Sex", "Pclass", "SibSp", "Survived"]]

corr_matrix = data_selected.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm_r", fmt=".2f", linewidths=0.5)

plt.title("Correlational Heatmap", fontsize=16)

plt.show()
```

•	Facet Grid: Create a facet grid of histograms showing age distribution separately for each class (Pclass). 

```python
facetGrid = sns.FacetGrid(data, col="Pclass", col_wrap=3, height=4, aspect=1.2)

facetGrid.map(sns.histplot, "Age", kde=True, color="blue", bins=10)

facetGrid.set_axis_labels("Age", "Frequency")
facetGrid.set_titles("Class {col_name}")

plt.subplots_adjust(top=0.85)
facetGrid.fig.suptitle("Age Distribution by Passenger Class (Pclass)", fontsize=16)
plt.show()
```

•	Pairplot: Generate a pairplot to analyze numerical relationships between Age, Fare, and Survived. Use “coolwarm” palette.

```python
sns.pairplot(data, vars=["Age", "Fare"], hue="Survived", palette="coolwarm", diag_kind="kde", height=3)
plt.suptitle("Pairplot of Age, Fare, and Survived", y=1.02, fontsize=16)
plt.show()
```
